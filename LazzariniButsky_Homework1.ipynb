Kepler's Equation:
M = E-esinE

```julia
function Kepler(M, e)
    E_old = M
    epsilon = 1.0
    while epsilon > 1.0E-12
        g = E_old - e*sin(E_old) - M
        g_prime = 1.0 - e*cos(E_old)
        E_new = E_old - (g/g_prime)
        epsilon = abs(E_old - E_new)/E_old
        E_old = E_new
    end
    return E_old
end
```

```julia
e = [0.0 0.25 0.5 0.75]
M = linspace(0,2.0*pi, 100)
E = Array{Float64}(100)
color = ["red", "green", "blue", "black"]
label = ["0.0", "0.25", "0.5", "0.75"]
for j = 1:4
    for i = 1:100
        E[i] = Kepler(M[i],e[j])
    end
    scatter(M, E, color=color[j], label=label[j])
end
legend(loc=4)
xlabel("M")
ylabel("E")
title("Solutions to Kepler's Equation")
```
